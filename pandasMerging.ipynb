```python
# right join
temp_df = pd.DataFrame({
    'student_id':[26,27,28],
    'name':['Nitish','Ankit','Rahul'],
    'partner':[28,26,17]
})

students = pd.concat([students,temp_df],ignore_index=True)
```

```python
students.merge(regs,how='right',on='student_id')
```

```python
# outer join
students.merge(regs,how='outer',on='student_id').tail(10)
```

```python
# 1. find total revenue generated
regs.merge(courses,how='inner',on='course_id')['price'].sum()
```

```python
# 2. find month by month revenue
temp_df = pd.concat([nov,dec],keys=['nov','dec']).reset_index()
temp_df
temp_df.merge(courses,how='inner',on='course_id').groupby(['level_0'])['price'].sum()
```

```python
# 3. Print the registration table
# cols -> name -> course -> price
regs.merge(courses,how='inner',on='course_id').merge(students,how='inner',on='student_id')[['name','price','course_name']]
```

```python
# 4. Plot bar chart for revenue/course
regs.merge(courses,how='inner',on='course_id').groupby('course_name')['price'].sum().plot(kind='bar')
```

```python
# 5. find students who enrolled in both the months
id = np.intersect1d(nov['student_id'],dec['student_id'])
```

```python
students[students['student_id'].isin(id)]
```

```python
# 6. find course that got no enrollment
# courses['course_id']
# regs['course_id'].unique().sort()

id = np.setdiff1d(courses['course_id'] ,regs['course_id'])
id
courses[courses['course_id'].isin(id)]
```

```python
# 7. find students who did not enroll into any courses

student_list = np.setdiff1d(students['student_id'],regs['student_id'])
student_list
students[students['student_id'].isin(student_list)].shape
```

```python
# 8. Print student name -> partner name for all enrolled students
# self join
students.merge(students,how='inner',left_on='partner',right_on='student_id')[['name_x','name_y']]
```

```python
# 9. find top 3 students who did most number enrollments
regs.merge(students,how='inner',on='student_id').groupby(['student_id','name'])['name'].count().sort_values(ascending=False).head(3)
```

```python
# 10. find top 3 students who spent most amount of money on courses
regs.merge(students,on='student_id').merge(courses,on='course_id').groupby(['student_id','name'])['price'].sum().sort_values(ascending=False).head(3)
```
